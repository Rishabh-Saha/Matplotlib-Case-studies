```python
import urllib.request

def read_text_from_url(url):
    try:
        response = urllib.request.urlopen(url)
        text = response.read().decode('utf-8')
        return text
    except urllib.error.URLError as e:
        return f"Error opening URL: {e}"
    except Exception as e:
        return f"An error occurred: {e}"

url = "https://cdn.upgrad.com/uploads/production/83a0597f-1dd2-4653-8194-cbef1870e5fb/data.txt"
text_content = read_text_from_url(url)

if "Error" not in text_content:
    print(text_content)
else:
    print(text_content)
```

```python
import spacy
```

```python
model = spacy.load("en_core_web_sm")
```

```python
list_of_sentences = text_content.split('.')
```

```python
money = []
for sentence in list_of_sentences:
    doc = model(sentence)
    # for token in doc:
    #     print(token.text, "--", token.pos_)
    for ent in doc.ents:
        print(ent.text, "--", ent.label_)
        if ent.label_ == "MONEY":
            money.append(int(ent.text))
```

```python
sum(money)/len(money)
```
